```
import pandas as pd 
import seaborn as sns 
sns.set()
```

```
train_df=pd.read_excel('Data_Train.xlsx')
test_df=pd.read_excel('Test_set.xlsx')
```

```
train_df.head()
```

```
combained_df = train_df.append(test_df, sort=False)
```

```
combained_df.head()
```

```
combained_df.dtypes
```

```
# Feature Engineering

def split_date_to_int(df):
  date_parse = df['Date_of_Journey'].str.split('/')
  return date_parse.str[0].astype(int), date_parse.str[1].astype(int), date_parse.str[2].astype(int)

combained_df['Day'], combained_df['Month'], combained_df['Year'] = split_date_to_int(combained_df)
```

```
combained_df.head()
```

```
combained_df.tail()
```

```
combained_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```
combained_df.head()
```

```
''' 
  Process the Arival_Time
'''
combained_df['Arrival_Time'] = combained_df['Arrival_Time'].str.split(' ').str[0]
combained_df.head()
```

```
''' 
  Process the Total_stops
'''
combained_df[combained_df['Total_Stops'].isnull()]
```

```
combained_df['Total_Stops'] = combained_df['Total_Stops'].fillna('1 stop')
```

```
combained_df['Total_Stops'] = combained_df['Total_Stops'].replace('non-stop', '0 stop')
```

```
combained_df['Stop'] = combained_df['Total_Stops'].str.split(' ').str[0].astype(int)
combained_df.drop('Total_Stops', axis=1, inplace=True)
```

```
combained_df.head()
```

```
''' 
  Arrival_Time Preprocess
'''
combained_df['Arrival_Hours'], combained_df['Arrival_Minutes'] = combained_df['Arrival_Time'].str.split(':').str[0].astype(int), combained_df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```
combained_df.head()
```

```
''' 
Depature Time Preprocess
'''
combained_df['Depature_Hours'], combained_df['Depature_Minutes'] = combained_df['Dep_Time'].str.split(':').str[0].astype(int), combained_df['Dep_Time'].str.split(':').str[1].astype(int)
```

```
combained_df.head()
```

```
combained_df.drop(['Arrival_Time', 'Dep_Time'], axis=1, inplace=True)
```

```
combained_df.head()
```

```
''' 
  Route Preprocess
'''
def route_preprocess_split(df):
  route_p = df['Route'].str.split('→ ')
  return route_p.str[0], route_p.str[1], route_p.str[2], route_p.str[3], route_p.str[4]

combained_df['Route_1'], combained_df['Route_2'], combained_df['Route_3'], combained_df['Route_4'], combained_df['Route_5'] = route_preprocess_split(combained_df)
```

```
combained_df.head()
```

```
combained_df['Route_1'].fillna("None", inplace=True)
combained_df['Route_2'].fillna("None", inplace=True)
combained_df['Route_3'].fillna("None", inplace=True)
combained_df['Route_4'].fillna("None", inplace=True)
combained_df['Route_5'].fillna("None", inplace=True)
```

```
combained_df.head()
```

```
combained_df.drop('Route', axis=1, inplace=True)
```

```
combained_df.head()
```

```
''' 
  Price Preprocess
'''
combained_df[combained_df['Price'].isnull()]
```

```
combained_df['Price'].fillna(combained_df['Price'].mean(), inplace=True)
```

```
combained_df.head()
```

```
combained_df.drop('Duration', axis=1, inplace=True)
```

```
combained_df.head()
```

```
combained_df.isnull().sum()
```

```
newDF = combained_df[["Airline", "Source", "Destination", "Day", "Month", "Year", "Arrival_Hours", "Arrival_Minutes", "Depature_Hours", "Depature_Minutes", "Route_1", "Route_2", "Route_3", "Route_4", "Route_5","Additional_Info","Price"]]
```
